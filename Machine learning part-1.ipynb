# machine learning part-1


**1. What is a parameter?**

A parameter is a configuration variable that is internal to the model and whose value can be estimated from the training data.

**2. What is correlation?**

Correlation measures the strength and direction of a linear relationship between two variables.

**3. What does negative correlation mean?**

Negative correlation means that as one variable increases, the other decreases. The correlation coefficient lies between -1 and 0.

**4. Define Machine Learning. What are the main components in Machine Learning?**
Machine Learning is a field of artificial intelligence that uses algorithms to allow computers to learn from data. Main components: data, model, algorithm, loss function, optimizer.

**5. How does loss value help in determining whether the model is good or not?**
Loss measures how far the model's prediction is from the actual target. A smaller loss indicates a better model.

**6. What are continuous and categorical variables?**

Continuous variables take infinite values within a range (e.g., height, weight). Categorical variables have discrete categories (e.g., gender, color).

**7. How do we handle categorical variables in Machine Learning? What are the common techniques?**

Using encoding techniques like one-hot encoding, label encoding, or ordinal encoding.

**8. What do you mean by training and testing a dataset?**

Training data is used to fit the model; testing data is used to evaluate the model's performance.

**9. What is sklearn.preprocessing?**

A module in scikit-learn for preprocessing data, including scaling, encoding, and normalization.

**10. What is a Test set?**

A subset of the dataset used to test the model after training to evaluate its performance on unseen data.

**11. How do we split data for model fitting (training and testing) in Python?**
Using `train_test_split()` from `sklearn.model_selection`.

**12. How do you approach a Machine Learning problem?**

Understand the problem, collect and clean data, perform EDA, preprocess, select and train model, evaluate, and tune.

**13. Why do we have to perform EDA before fitting a model to the data?**

EDA helps understand the data, identify patterns, outliers, missing values, and determine appropriate preprocessing steps.

**16. How can you find correlation between variables in Python?**

Using `.corr()` method in pandas.

**17. What is causation? Explain difference between correlation and causation with an example.**

Causation implies that one variable causes the change in another. Correlation only shows a relationship. E.g., Ice cream sales and drowning deaths are correlated but not causally linked.

**18. What is an Optimizer? What are different types of optimizers? Explain each with an example.**

Optimizers adjust model parameters to minimize the loss. Types:
- Gradient Descent
- Stochastic Gradient Descent (SGD)
- Adam
Example: `optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)`

**19. What is sklearn.linear_model?**

A module in scikit-learn with linear models like LinearRegression, LogisticRegression.

**20. What does model.fit() do? What arguments must be given?**

Trains the model. Requires training data and target variable, e.g., `model.fit(X_train, y_train)`

**21. What does model.predict() do? What arguments must be given?**

Predicts target values using the trained model. Requires input data, e.g., `model.predict(X_test)`

**22. What is feature scaling? How does it help in Machine Learning?**

Scaling transforms features to a similar range. It helps algorithms converge faster and perform better.

**23. How do we perform scaling in Python?**

Using StandardScaler or MinMaxScaler from `sklearn.preprocessing`.

**24. What is sklearn.preprocessing?**

(Repeated) It's for scaling, normalization, and encoding of data.

**25. How do we split data for model fitting (training and testing) in Python?**

(Repeated) Use `train_test_split()` function from `sklearn.model_selection`.

**26. Explain data encoding?**

Transforming categorical data into numeric form so it can be used in models. Common methods: one-hot, label, and ordinal encoding.
